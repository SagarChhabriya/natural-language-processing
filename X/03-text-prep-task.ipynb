```python
import requests as r 

response = r.request(method="GET",url="https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1")
response.json()
```

```python
import pandas as pd

df = pd.DataFrame(response.json()['results'])
df
```
